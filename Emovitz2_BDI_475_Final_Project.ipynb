# Aggregations 1 and 2

1. Pokemon that are only 1 type
2. Pokemon that are double type



```python
### 1 ###
df_only_type1 = df[df['Type 2'].isna()]
df_only_type1 = df_only_type1.drop(columns=['Type 2']) #inplace=True was giving a weird error

### 2 ###
df_two_types = df[df['Type 2'].notna()]
```

# Aggregation 3


* Number of different pokemon that are only 1 type by type and percentage of all 1 type pokemon




```python
### 3 ###
num_of_pokemon_by_type1 = df_only_type1.groupby('Type 1', as_index=False).agg({'Name':'count'}).rename(columns={'Name': 'Occurences'}).sort_values('Occurences', ascending=False)
num_of_pokemon_by_type1['Percentage'] = num_of_pokemon_by_type1['Occurences'] / num_of_pokemon_by_type1['Occurences'].sum() * 100
num_of_pokemon_by_type1
```

Of all Pokemon that are only one type, Normal is the most common type with Water being the second most common.

# Aggregation 4

* Different types of double type/most common double type

```python
#Sorted by Generation, Type 1, Type 2
df_two_types_sorted = df_two_types.sort_values(['Generation','Type 1', 'Type 2'], ascending=[True, True, True])
# df_two_types_sorted

### 4 ###
df_most_common_double_type = df_two_types_sorted.groupby(['Type 1', 'Type 2']).agg({'Name':'count'}).rename(columns={'Name':'Occurences'}).reset_index().sort_values('Occurences', ascending=False)
df_most_common_double_type
```

Of all Pokemon that are only double type, Normal and Flying is the most common double type with Grass and Poison being the second most double type common.

# Aggreagation 5
* Which generation has the most double types

```python
### 5 ###
df_double_type_by_gen = df_two_types_sorted.groupby('Generation').agg({'Name':'count'}).rename(columns={'Name':'Occurences'}).sort_values('Occurences', ascending=False)
df_double_type_by_gen
```

Generation 3 and 5 have the most Pokemon that are double types.

# Aggregation 6
* Percentage of pokemon that are double type by generation

```python
### 6 ###
df_total_pokemon_by_gen = df.groupby('Generation').agg({'Name':'count'}).rename(columns={'Name':'Occurences'})
df_double_type_by_gen['Total pokemon'] = df_total_pokemon_by_gen['Occurences']
df_double_type_by_gen['Percent Double Type'] = round(df_double_type_by_gen['Occurences'] / df_double_type_by_gen['Total pokemon'] * 100,2)
df_double_type_by_gen.sort_values('Percent Double Type', ascending=False, inplace = True)
df_double_type_by_gen
```

Generation 6 has the most Pokemon that are double type by percentage of all pokemon in that generation.

# Aggregation 7
* Best pokemon that are only 1 type determined by specified stat


```python
### 7 ###
def best_pokemon_by_stat(Stat, Legendary=True):

  if Legendary == True:
    df_type_and_stat = df_only_type1.groupby(['Type 1']).agg({Stat : 'max'}).sort_values(Stat, ascending=False)
    types = df_type_and_stat.index.tolist()
    stat_title = df_type_and_stat[Stat].tolist()

    p = []
    for i,k in zip(types, stat_title):
      ind = df_only_type1[(df_only_type1['Type 1'] == i) & (df_only_type1[Stat] == k)]
      p.append(ind.index[0])

    df = df_only_type1.loc[p].sort_values(Stat, ascending=False)

  else:
    df_no_leg = df_only_type1[df_only_type1['Legendary'] == False]

    df_type_and_stat = df_no_leg.groupby(['Type 1']).agg({Stat : 'max'}).sort_values(Stat, ascending=False)
    types = df_type_and_stat.index.tolist()
    stat_title = df_type_and_stat[Stat].tolist()

    p = []
    for i,k in zip(types, stat_title):
      ind = df_no_leg[(df_no_leg['Type 1'] == i) & (df_no_leg[Stat] == k)]
      p.append(ind.index[0])

    df = df_only_type1.loc[p].sort_values(Stat, ascending=False)
  
  return df
```

```python
best_type1_pokemon_by_stat = best_pokemon_by_stat('Defense', Legendary=False)
best_type1_pokemon_by_stat
```

Looking at Total and including Legendaries, Pyschic is the best 1 type Pokemon. Looking at Defense and excluding Legendaries, Steel is the best 1 type Pokemon. 

# Aggregation 8
* Generation with most legendaries by count and percentage of lengedaries

```python
### 8 ###
df_legendary = df[df['Legendary'] == True]
df_legendary_group = df_legendary.groupby('Generation', as_index=False).agg({'Name':'count'}).rename(columns={'Name':'Occurences'}).sort_values('Occurences', ascending = False)
df_legendary_group['Percentage'] = round(df_legendary_group['Occurences'] / df_legendary_group['Occurences'].sum() * 100,2)
df_legendary_group
```

Only considering legendary Pokemon in each generation, generation 3 has the most legendaries.

# Visualizations
These are used to easily show different aggregations.

```python
#Box plot of Total column for all pokemon
fig = px.box(df, x='Total', orientation='h', title='Distribution of Pokemon by Total column')
fig.show()
```

This shows the distribution of all Pokemon sorted by the Total Column.

```python
#Box plot of Total for all pokemon by Type 1
fig = px.box(df, x='Total', y='Type 1', orientation='h', title='Distribution of Pokemon by Total column separated by Type 1')
fig.show()
```

This shows the distribution of all Pokemon sorted by the Total column and by Type 1.

```python
#Sunburst of pokemon that have 2 types
fig = px.sunburst(
    df_most_common_double_type,
    path=['Type 1', 'Type 2'],
    values = 'Occurences',
    width = 800,
    height = 900,
    title = 'Most common Double types'
)

fig.show()
```

This shows only double type pokemon where the inner circle is the distribution of the primary type and the outer circle is the distribution of secondary type.

```python
#Bar graph of only 1 type pokemon
fig = px.bar(
    num_of_pokemon_by_type1,
    x = 'Type 1',
    y = 'Occurences',
    title = 'Number of Pokemon that are only 1 type'
)

fig.show()
```

This shows the amount of only one type Pokemon that occur.

```python
#Percentage of Legendaries by Generation
fig = px.pie(
    df_legendary_group,
    names = 'Generation',
    values = 'Percentage',
    width = 800,
    height = 700,
    title = 'Generation with most legendaries by percent'
)

fig.show()
```

This shows the percentage of legendaries per genertation of all the legendaries in the game.
